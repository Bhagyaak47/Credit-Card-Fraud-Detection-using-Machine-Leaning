```python
import numpy as np  # used to make arrays
import pandas as pd  # make structured data & preprocessing 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to pandas Dataframe
cc_data = pd.read_csv("creditcard.csv")
cc_data.head(5)
```

```python
cc_data.tail()
```

```python
#dataset info
cc_data.info()
```

```python
#checking number of missing values

cc_data.isnull().sum()
```

```python
# distribution of legit transaction

cc_data["Class"].value_counts()
```

This Dataset is highly unbalanced 
0 -> Normal Transaction
1 -> fraudant Transaction

```python
# separating the data for analysis
legit = cc_data[cc_data.Class == 0] 
fraud = cc_data[cc_data.Class == 1]
```

```python
legit.shape
fraud.shape
```

```python
#statistical measures of the data 
legit.Amount.describe()
```

```python
#compare the values for both transactions
cc_data.groupby("Class").mean()
```

Build a sample dataset containing similar distribution of normal transaction & fraudulent transaction 

Number of Fraudulant Transaction ----> 492


```python

legit_sample = legit.sample(n= 492)
```

Concatenating two Dataframes 

```python
new_datasets = pd.concat([legit_sample,fraud],axis = 0)
```

```python
new_datasets.tail()
```

```python
new_datasets["Class"].value_counts()
```

Spliiting the data into features & targets


```python
X = new_datasets.drop(columns ="Class",axis = 1)
Y = new_datasets["Class"]
```

```python
X
```

```python
Y
```

Split the data into Training data & testing data 

```python
X_train,X_test, Y_train,Y_test = train_test_split(X,Y,test_size=0.2, stratify=Y ,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training
Logistics Regression

```python
model = LogisticRegression()
```

```python
#training the logistic  Regression Model with Training Data
model.fit(X_train,Y_train)
```

```python
#Accuracy Score

X_train_prediction =model.predict(X_train)
trainining_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print("Accuracy on training data :",trainining_data_accuracy )
```

```python
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on test data :",test_data_accuracy )
```
